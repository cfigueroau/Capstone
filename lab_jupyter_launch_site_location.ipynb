```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import radians, cos, sin, asin, sqrt

# Función para calcular distancia en km entre dos coordenadas
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Radio de la Tierra en km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1 
    dlon = lon2 - lon1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    return R * c
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Coordenadas del punto más cercano en la costa (usadas como ejemplo)
coast_lat = 28.56367
coast_lon = -80.57163

# Obtener coordenadas del sitio de lanzamiento
launch_site = launch_sites_df[launch_sites_df['Launch Site'] == 'CCAFS LC-40'].iloc[0]
launch_lat = launch_site['Lat']
launch_lon = launch_site['Long']

# Calcular distancia
distance_coastline = calculate_distance(launch_lat, launch_lon, coast_lat, coast_lon)
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
# distance_marker = folium.Marker(
#    coordinate,
#    icon=DivIcon(
#        icon_size=(20,20),
#        icon_anchor=(0,0),
#        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
#        )
#    )

# Crear mapa centrado en el sitio de lanzamiento
coast_map = folium.Map(location=[launch_lat, launch_lon], zoom_start=14)

# Agregar marcador del sitio de lanzamiento
folium.Marker(
    [launch_lat, launch_lon],
    icon=folium.Icon(color='blue'),
    popup='Sitio de lanzamiento'
).add_to(coast_map)

# Agregar marcador de la costa
folium.Marker(
    [coast_lat, coast_lon],
    icon=folium.Icon(color='gray'),
    popup='Punto en línea costera'
).add_to(coast_map)

# Agregar texto con la distancia usando DivIcon
folium.Marker(
    location=[(launch_lat + coast_lat)/2, (launch_lon + coast_lon)/2],
    icon=folium.DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html=f'<div style="font-size: 12pt; color:#d35400;"><b>{distance_coastline:.2f} KM</b></div>'
    )
).add_to(coast_map)

# Agregar línea entre el sitio y la costa
folium.PolyLine(
    [[launch_lat, launch_lon], [coast_lat, coast_lon]],
    color='red',
    weight=3
).add_to(coast_map)

# Mostrar el mapa
coast_map
```
